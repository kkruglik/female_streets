```python
import pandas as pd
pd.set_option('display.max_rows', 1000)
```

Считаем отдельные фамилии в датасете

```python
add_names = [
    "крупской",
    "крупская",
    "люксембург",
    "цеткин",
    "космодемьянской",
    "космодемьянская",
    "ульяновой",
    "осипенко",
    "гризодубова",
    "гризодубовой",
    "раскова",
    "расковой",
    "терешкова",
    "терешковой"
]
```

```python
# загрузка датасета
dataset = pd.read_csv("administrative_addresses.csv", dtype="str", sep=";")

dataset = dataset[dataset.street.notna()]
dataset = dataset.filter(items=['region', 'settlement', 'street', 'street_type'])
dataset = dataset.drop_duplicates(subset=['region', 'settlement', 'street', 'street_type'])
dataset.street = dataset.street.str.lower()
```

Какие есть вариации улиц с этими фамилиями

```python
mask = dataset['street'].str.contains('|'.join(add_names))
dataset[mask].groupby("street").size().sort_values(ascending=False).to_frame()
```

```python
dataset['name'] = None

for name in add_names:
    mask = dataset['street'].str.contains(name, regex=False, case=False)
    dataset.loc[mask, "name"] = name
```

```python
dataset.loc[dataset.name == "космодемьянская", "name"] = "космодемьянской"
dataset.loc[dataset.name == "крупская", "name"] = "крупской"
dataset.loc[dataset.name == "терешкова", "name"] = "терешковой"
dataset.loc[dataset.name == "раскова", "name"] = "расковой"
```

Сколько улиц с этими фамилиями в датасете

```python
mask = dataset['street'].str.contains('|'.join(add_names))
dataset[mask].groupby("name").size().sort_values(ascending=False).to_frame()
```

Сколько таких улиц по регионам

```python
dataset[mask].groupby(["region", "name"]).size().to_frame()
```
